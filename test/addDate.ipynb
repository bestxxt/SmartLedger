```python
import requests
import random
from datetime import datetime, timedelta
from faker import Faker

fake = Faker()
```

```python

fake = Faker()

# Configuration
API_BASE = "http://localhost:3000"
LOGIN_URL = f"{API_BASE}/api/account/login"
TX_URL    = f"{API_BASE}/api/app/transactions"

# Replace with valid user credentials
CREDENTIALS = {
    "email": "admin",
    "password": "password"
}

# Start a session and login to obtain cookies
session = requests.Session()
login_resp = session.post(LOGIN_URL, json=CREDENTIALS)
if not login_resp.ok:
    raise Exception(f"Login failed ({login_resp.status_code}): {login_resp.text}")
print("Logged in successfully, cookies:", session.cookies.get_dict())
```

```python
# Category definitions
CATEGORIES = {
    "Food":        ["Groceries", "Restaurants", "Cafe", "Snacks", "Takeout"],
    "Transport":   ["Bus", "Train", "Taxi", "RideShare", "Fuel", "Parking"],
    "Shopping":    ["Clothes", "Electronics", "Books", "Gifts", "Home", "Beauty"],
    "Entertainment":["Movies", "Music", "Games", "Events", "Travel"],
    "Health":      ["Medicine", "Gym", "Doctor", "Supplements"],
    "Bills":       ["Rent", "Utilities", "Internet", "Phone", "Insurance"],
    "Income":      ["Salary", "Bonus", "Interest", "Refund"],
    "Other":       ["Miscellaneous"],
}
CURRENCIES = ["USD", "CAD", "EUR", "GBP"]
```

```python
# Generate and post 100 random transactions
for i in range(1, 101):
    tx_type = random.choice(["income", "expense"])
    category = random.choice(list(CATEGORIES.keys()))
    subcat = random.choice(CATEGORIES[category])
    amount = round(
        random.uniform(5, 500) if tx_type == "expense"
        else random.uniform(100, 5000),
        2
    )
    ts = datetime.now() - timedelta(
        days=random.randint(0, 365),
        hours=random.randint(0, 23),
        minutes=random.randint(0, 59),
        seconds=random.randint(0, 59),
    )

    payload = {
        "amount": amount,
        "type": tx_type,
        "category": category,
        "subcategory": subcat,
        "timestamp": ts.isoformat() + "Z",
        "note": fake.sentence(nb_words=6),
        "currency": random.choice(CURRENCIES),
        "tags": fake.words(nb=random.randint(0, 3)),
        "location": fake.city(),
    }

    resp = session.post(TX_URL, json=payload)
    if resp.ok:
        print(f"[{i:03}] ✔ {category}/{subcat} @ {ts.strftime('%Y-%m-%d %H:%M')} → {resp.status_code}")
    else:
        print(f"[{i:03}] ✘ Failed ({resp.status_code}): {resp.text}")
```
